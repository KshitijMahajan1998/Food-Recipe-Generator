```python
train_data['Keywords'] = train_data['Keywords'].astype(str)
train_data['RecipeIngredientQuantities'] = train_data['RecipeIngredientQuantities'].astype(str)
train_data['RecipeIngredientParts'] = train_data['RecipeIngredientParts'].astype(str)
```

```python
import re

def map_str_to_list(string):
    #pattern = re.compile(r'\"(.+)\"')
    pattern = re.compile(r'\"([^"]+)\"')
    return pattern.findall(string)

def map_for_series(series: pd.Series):
    return series.apply(lambda i: map_str_to_list(i))
```

```python
modify_col_list = ['Keywords', 'RecipeIngredientQuantities', 'RecipeIngredientParts']
train_data = train_data.apply(lambda x: map_for_series(x) if x.name in modify_col_list else x)
```

```python
train_data["No_of_Ingredients"] = train_data["RecipeIngredientParts"].apply(lambda x: len(x))
```

```python
receipe_data = pd.read_csv("/content/gdrive/MyDrive/MSBA/AML_Project /recipes.csv")
```

```python
receipe_data.head()
```

```python
final_data = train_data.merge(receipe_data[['RecipeServings', 'RecipeYield', 'RecipeId']], how = 'left', on = 'RecipeId')
```

```python
final_data['RecipeServings'].fillna(final_data['RecipeServings'].min(), inplace=True)
final_data.head()
```

```python
Calorie_Reference = 2000
RDV_Fat_Content = 78
RDV_sat_fat_content = 20
RDV_cholestrol_content = 300
RDV_sodium_content = 2300
RDV_carb_content = 275
RDV_fiber = 28
RDV_sugar = 200
RDV_protein = 50
```

```python
final_data.columns
```

```python
final_data['per_fat_DV'] = (final_data['FatContent'] / RDV_Fat_Content)
final_data['per_satfat_DV'] = (final_data['SaturatedFatContent'] / RDV_sat_fat_content)
final_data['per_chol_DV'] = (final_data['CholesterolContent'] / RDV_cholestrol_content)
final_data['per_sodium_DV'] = (final_data['SodiumContent'] /RDV_sodium_content)
final_data['per_carb_DV'] = (final_data['CarbohydrateContent'] / RDV_carb_content)
final_data['per_fiber_DV'] = (final_data['FiberContent'] / RDV_fiber)
final_data['per_sugar_DV'] = (final_data['SugarContent'] / RDV_sugar)
final_data['per_prot_DV'] = (final_data['ProteinContent'] / RDV_protein)
```

```python
final_data.head()
```

```python
final_data['per_fat_DV'] = (final_data['per_fat_DV']) / (final_data['RecipeServings'])
final_data['per_satfat_DV'] = (final_data['per_satfat_DV']) / (final_data['RecipeServings'])
final_data['per_chol_DV'] = (final_data['per_chol_DV']) / (final_data['RecipeServings'])
final_data['per_sodium_DV'] = (final_data['per_sodium_DV']) / (final_data['RecipeServings'])
final_data['per_carb_DV'] = (final_data['per_carb_DV']) / (final_data['RecipeServings'])
final_data['per_fiber_DV'] = (final_data['per_fiber_DV']) / (final_data['RecipeServings'])
final_data['per_sugar_DV'] = (final_data['per_sugar_DV']) / (final_data['RecipeServings'])
final_data['per_prot_DV'] = (final_data['per_prot_DV']) / (final_data['RecipeServings'])
```

```python
def condition(row):
  if (row['per_satfat_DV'] > 0.20 or row['per_sodium_DV'] > 0.20 or row['per_sugar_DV'] > 0.20):
    return 'Not a Nutritious Meal'
  elif (row['per_fiber_DV'] < 0.05 or row['per_prot_DV'] < 0.05): 
    return 'Not a Nutritious Meal'
  elif (row['per_fat_DV'] < 0.20 and row['per_chol_DV'] < 0.20 and row['per_carb_DV']  < 0.20): 
    return 'Balanced Nutritious Meal'
  else: 
    return 'Unbalanced Nutritious Meal'
```

```python
final_data['Nutrition_Assessment'] = final_data.apply(lambda row: condition(row), axis=1)
```
